```python
import pynbody 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

def patch_asscalar(a):
    return a.item()

def fourier(x,y):
    theta=np.arctan2(y,x)
    a2=np.sum(np.cos(2*theta))
    b2=np.sum(np.sin(2*theta))
    
    f=np.sqrt(a2*a2+b2*b2)/len(x)
    phi=np.arctan2(b2,a2)
    #print(np.sqrt(a2*a2+b2*b2))
    l=[abs(f),phi, len(x)]
    return l
```

# Model D0

```python
bar_len=[]
for i in range (499,500):
    if i< 10 :
        snap='snapshot_00'+str(i)
    elif i<100:
        snap='snapshot_0'+str(i)
    else :
        snap='Spin100_D0/snapshot_'+str(i)
    s=pynbody.load(snap)

    
    #s=pynbody.load('Spin000/snapshot_500')
    disk=s.star
    x=disk['pos'][:,0]-np.mean(disk['pos'][:,0])
    y=disk['pos'][:,1]-np.mean(disk['pos'][:,1])
    z=disk['pos'][:,2]-np.mean(disk['pos'][:,2])
    r=np.sqrt(x**2+y**2)
    
    A2byA0=[]
    bin=80
    rmax=20.0
    A2_tot=[]
    phi=[]
    A0=[]
    
    for i in range(0,bin):
        binsize=rmax/bin
        msk=np.where((r>i/4.0)&(r<(i/4.0+binsize)))
        x0=x[msk]
        y0=y[msk]
        z0=z[msk]
        
        A2byA0.append(fourier(x0,y0)[0])
        phi.append(fourier(x0,y0)[1])
        
        
        #msk2=np.where(r<i/4.0)
        #x1=x[msk2]
        #y1=y[msk2]
        #z1=z[msk2]
        
        #A2_tot.append(fourier(x1,y1)[0])   
        #A0.append(fourier(x1,y1)[2]) 
    c=np.array(A2byA0)
    l=np.where(c==max(c)) #indices of maximum A2byA0
    k=np.where(c < max(c)*0.60)    #indices where
    bar_len.append(k[0][k[0]>l[0]][0]*binsize) # only taking first indices where 60% of maximum value is reached after peak.
     

print(bar_len, 'kpc')
```
